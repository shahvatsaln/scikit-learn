```python
import numpy as np # linear algebra
import pandas as pd
import re # text processing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_validate,cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
train_data=pd.read_csv("/home/ubuntu/ML/titanic/train.csv")
test_data=pd.read_csv("/home/ubuntu/ML/titanic/test.csv")
```

```python
#train_data.shape#891,12

test_data.shape#418,11
```

```python
train_data.head()

X_train = train_data.drop(['Survived'],axis=1)
y_train = train_data['Survived']

X_test=test_data
```

```python
sex_encoder = LabelEncoder()
X_train['Sex'] = sex_encoder.fit_transform(X_train['Sex'])
X_test['Sex'] = sex_encoder.fit_transform(X_test['Sex'])
```

```python
X_train['Has_Cabin']=X_train['Cabin'].apply(lambda x: 0 if type(x) == float else 1)
#X_train['Sex']=X_train['Sex'].map({"male":1,"female":0})
X_train['Embarked']=X_train['Embarked'].map({"S":2,"C":1,"Q":3}).fillna(0).astype(int)


X_test['Has_Cabin']=X_test['Cabin'].apply(lambda x: 0 if type(x) == float else 1)
#X_test['Sex']=X_test['Sex'].map({"male":1,"female":0})
X_test['Embarked']=X_test['Embarked'].map({"S":2,"C":1,"Q":3}).fillna(0).astype(int)
```

```python
#X_train['Age'].isnull().sum()
X_train['Age']=X_train['Age'].fillna(X_train['Age'].mean()).astype(int)
#X_train
#X_train['Age'].mean()

X_test['Age']=X_test['Age'].fillna(X_test['Age'].mean()).astype(int)
#X_test
```

```python
X_train=X_train.drop(['PassengerId','Name','Ticket','Fare','Cabin'],axis=1)
X_test=X_test.drop(['PassengerId','Name','Ticket','Fare','Cabin'],axis=1)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_test=lr.predict(X_test)


#logreg_acc_score = round(logreg.score(X_train, y_train) * 100, 2)
```
